```python
import pandas as pd
# import streamlit as st
import numpy as np
```

```python
data = pd.read_csv('https://privacyrights.org/sites/default/files/2020-01/PRC%20Data%20Breach%20Chronology%20-%201.13.20.csv',usecols=['Date Made Public', 'Company', 'City', 'State', 'Type of breach',
       'Type of organization', 'Total Records', 'Description of incident',
       'Year of Breach', 'Latitude',
       'Longitude'])
```

```python
# data.head()

Year = data['Year of Breach'].tolist()
Date = data['Date Made Public'].tolist()

Difference=0
for i in range(len(data)):
    Temp = Date[i].split('/')
    if int(Temp[2].strip())==Year[i]:
        continue
    else:
        Difference+=1

# print(Difference/len(data))

# There are differences between Data Made Public and Data of Breach; 3.73% of instances.

data['Date Made Public']=pd.to_datetime(data['Date Made Public'])
data['Total Records']=data['Total Records'].str.replace(',','').fillna('0')
data['Total Records']=data['Total Records'].astype('int64')
data.dtypes
```

```python
# data.head()

City = data['City'].to_list()
State = data['State'].to_list()
Longitude = data['Longitude'].to_list()
Latitude = data['Latitude'].to_list()

badrows=[]
for i in range(len(data)):
    if pd.isna(City[i]) and pd.isna(State[i]):
        if pd.isna(Longitude[i]) and pd.isna(Latitude[i]):
            badrows.append(i)
        else:
            continue
    else:
        continue

# print(len(badrows))
# print(badrows)
```

```python
# Fill in 'Type of Breach' Column
TOB_Dict = {
'CARD':'Credit/Debit Card Fraud',
# Fraud Involving Debit and Credit Cards Not Via Hacking (skimming devices at point-of-service terminals, etc.)

'HACK':'Hacker/Malware',
# Hacked by an Outside Party or Infected by Malware

'INSD' : 'Insider Employee/Contractor/Customer',
# Insider (employee, contractor or customer)

'PHYS' : 'Physical Breach - Paper',
# Physical (paper documents that are lost, discarded or stolen)

'PORT' : 'Physical Breach - Device',
# Portable Device (lost, discarded or stolen laptop, PDA, smartphone, memory stick, CDs, hard drive, data tape, etc.)

'STAT' : 'Physical Breach - Hardware',
# Stationary Computer Loss (lost, inappropriately accessed, discarded or stolen computer or server not designed for mobility)

'DISC' : 'Unintended Disclosure',
# Unintended Disclosure Not Involving Hacking, Intentional Breach or Physical Loss (sensitive information posted publicly, mishandled or sent to the wrong party via publishing online, sending in an email, sending in a mailing or sending via fax) 

'UNKN' : 'Unknown'}
# Unknown (not enough information about breach to know how exactly the information was exposed)

data = data.replace(TOB_Dict)
```

```python
# BSF
# Businesses (Financial and Insurance Services)
data = data.replace('BSF','Financial/Insurance Business')

# BSO
# Businesses (Other)
data = data.replace('BSO','Other Business')

# BSR
# Businesses (Retail/Merchant including Online Retail)
data = data.replace('BSR','Retail Business')

# EDU
# Educational Institutions
data = data.replace('EDU','Educational Institution')

# GOV
# Government & Military
data = data.replace('GOV','Government/Military')

# MED
# Healthcare, Medical Providers and Medical Insurance Services
data = data.replace('MED','Healthcare/Medical Services')

# NGO
# Nonprofits
data = data.replace('NGO','Nonprofit')

# UNKN
# Unknown
data = data.replace('UNKN','Unknown')

data.head()
```

```python
data.to_csv('PRC_Clean.csv')

# pd.isna(data['Longitude']).sum()
# 2474 rows of data are not be accoutned for.
```
